```python
# outliers
def number_of_outliers(df):
    
    df = df.select_dtypes(exclude = 'object')
    
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    
    return ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
```

```python
# There are NO outliers... Just need to drop non 0 and 1 values.

total = number_of_outliers(df)
total
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model = LogisticRegression()

model.fit(x_train,y_train)
y_pred_baseline = model.predict(x_test)
print('Acurracy of train ',(model.score(x_train, y_train)*100))
print('Acurracy of test ',(accuracy_score(y_test, y_pred_baseline)*100))
```

```python
cm1 = confusion_matrix(y_test, y_pred_baseline)

f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm1,annot = True,cmap='Reds',linewidths=1,linecolor='k',mask=False,fmt = ".0f",cbar=True)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```
